```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import scipy
import math
```

```python
data = pd.read_csv('/home/suraj/Desktop/fods_1.csv')
```

```python
data.head()
```

```python
def f1(x, y):
  Y = 0
  for i in data.values:
    Y += 0.5*(np.square(x*i[0] + y*i[1] - i[2]))
  return Y
```

```python
w1 = np.linspace(-4,4, 100)
w2 = np.linspace(-4,4, 100)
W1, W2 = np.meshgrid(w1, w2)
Y = f1(W1,W2)
```

```python
cm = plt.cm.get_cmap('viridis')
plt.scatter(W1, W2, c=f1(W1,W2), cmap=cm)
plt.show()
```

```python
cp = plt.contour(W1, W2, f1(W1,W2), levels=40)
plt.clabel(cp, inline=1, fontsize=5)
plt.xlabel('W1')
plt.ylabel('W2')
plt.show()
```

```python
def plot(n1,n2,q,r):
    N1, N2 = np.meshgrid(n1, n2)
    cp = plt.contour(N1, N2, f1(N1,N2),linestyles='dashed', levels=40)
    Y = np.abs(np.power(N1,q))+np.abs(np.power(N2,q))
    cp = plt.contour(N1, N2, Y, levels=[r],colors='r')
    plt.clabel(cp, inline=1, fontsize=9)
    plt.xlabel('W1')
    plt.ylabel('W2')
    plt.show()
```

```python
q=2
r=1.4
n1 = np.linspace(0,1.5, 50)
n2 = np.linspace(0,1.5, 50)
plot(n1,n2,q,r)
```
